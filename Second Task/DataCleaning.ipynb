```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
# Handling Missing Values
# Fill missing categorical values with 'Unknown'
categorical_columns = ['Brand', 'VehicleModel', 'Type', 'color', 'fuel', 'type of gear']
df[categorical_columns] = df[categorical_columns].fillna('Unknown')
```

```python
df.isna().sum()
```

```python
# Fill missing numeric columns with median
df['rating'] = df['rating'].fillna(df['rating'].median())
df['ManufacturingYear'] = df['ManufacturingYear'].fillna('Unknown')
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
# Extract numeric data from CylinderCount
df['CylinderCount'] = pd.to_numeric(df['CylinderCount'].str.extract('(\d+)')[0], errors='coerce').fillna(0)
```

```python
df['CylinderCount']
```

```python
# Extract numeric data from #airbags
df['#airbags'] = pd.to_numeric(df['#airbags'].str.extract('(-?\d+)')[0], errors='coerce').fillna(0)
```

```python
df['#airbags']
```

```python
# Extract numeric data from Odometer
df['Odometer'] = pd.to_numeric(df['Odometer'].str.replace(r'\D', '', regex=True), errors='coerce').fillna(0)
```

```python
df['Odometer']
```

```python
# Duty Column
df['Duty'] = pd.to_numeric(df['Duty'].str.replace(r'[^\d.]', '', regex=True), errors='coerce').fillna(0)
```

```python
df['Duty']
```

```python
# Convert ManufacturingYear to numeric
df['ManufacturingYear'] = pd.to_numeric(df['ManufacturingYear'], errors='coerce').fillna(0).astype(int)
```

```python
df['ManufacturingYear']
```

```python
df.head()
```

```python
# Brand Column
def clean_brand(brand):
    if pd.isna(brand) or brand.strip() == '':
        return 'Unknown'
    brand = brand.lower()  # Convert to lowercase
    brand = re.sub(r'i love |yes you got it|is the best|i own ', '', brand)
    brand = re.sub(r'[^a-z\s]', '', brand)
    brand = brand.strip()

    brand = brand.split()[0] if brand else 'Unknown'
    return brand.title()

df['Brand'] = df['Brand'].apply(clean_brand)
```

```python
df['Brand']
```

```python
df.head()
```

```python
# Clean and standardize Type
def clean_type(type_value):
    if pd.isna(type_value) or type_value.strip() == '':
        return 'Unknown'
    type_value = type_value.lower()
    type_value = re.sub(r'[^a-z\s]', '', type_value)
    type_value = re.sub(r'\b(woow|super|mega|ultra)\b', '', type_value)
    type_value = type_value.strip()

    type_value = type_value.split()[0] if type_value else 'Unknown'
    return type_value.title()

df['Type'] = df['Type'].apply(clean_type)
```

```python
df.head()
```

```python
# Clean and standardize VehicleModel
def clean_vehicle_model(model):
    if pd.isna(model) or model.strip() == '':
        return 'Unknown'
    model = model.lower()
    model = re.sub(r'\b(the|right|its|i have|yes you got it|i own|fastest model ever|Its|My|Model|Which|Is|Is  Best|Its|My Model Which Is|Is Best|Touring)\b', '', model)
    model = re.sub(r'[^a-z0-9\s]', '', model)
    model = model.strip()
    return model.title() if model else 'Unknown'

df['VehicleModel'] = df['VehicleModel'].apply(clean_vehicle_model)
```

```python
df['VehicleModel']
```

```python
df['#airbags'] = pd.to_numeric(df['#airbags'], errors='coerce').fillna(0)

# Cap the maximum value at 16
df['#airbags'] = df['#airbags'].apply(lambda x: min(x, 16))
```

```python
non_negative_mean = int(df.loc[df["#airbags"] >= 0, "#airbags"].mean())
df["#airbags"] = df["#airbags"].apply(lambda x: non_negative_mean if x < 0 else x)
```

```python
df.head()
```

```python
df = df.drop(columns='ID')
```

```python
df.head()
```

```python
# Put zero values in ManufacturingYear col to median
median_year = df.loc[df["ManufacturingYear"] > 0, "ManufacturingYear"].median()
df["ManufacturingYear"] = df["ManufacturingYear"].apply(lambda x: median_year if x == 0 else x)
```

```python
df.head()
```

```python
df['fuel'].value_counts()
```

```python
def clean_fuel(fuel_type):
    fuel_type = fuel_type.strip().lower()
    if fuel_type in ["dies", "diesel"]:
        return "Diesel"
    elif fuel_type in ["hyb", "electric hybrid"]:
        return "Hybrid"
    elif fuel_type in ["other", "others"]:
        return "Other"
    elif fuel_type == "natural gas":
        return "Natural Gas"
    elif fuel_type == "gas":
        return "Gas"
    elif fuel_type == "unknown":
        return "Unknown"
    else:
        return fuel_type.capitalize()

df["fuel"] = df["fuel"].apply(clean_fuel)
```

```python
df['fuel'].value_counts()
```

```python
df.head()
```

```python
df['CylinderCount'].value_counts()
```

```python
median_cylinders = df.loc[df["CylinderCount"] > 0, "CylinderCount"].median()
df["CylinderCount"] = df["CylinderCount"].replace(0, median_cylinders)

# Group rare CylinderCount values into "Other"
rare_threshold = 5
value_counts = df["CylinderCount"].value_counts()
df["CylinderCount"] = df["CylinderCount"].apply(
    lambda x: "Other" if value_counts[x] < rare_threshold else x
)
```

```python
df['CylinderCount'].value_counts()
```

```python
df.head()
```

```python
mean_price_above_5000 = df.loc[df["price"] >= 5000, "price"].mean()

df["price"] = df["price"].apply(lambda x: mean_price_above_5000 if x < 5000 else x).astype(int)
```

```python
df['price']
```

```python
df["type of gear"] = df["type of gear"].str.lower()

df["type of gear"] = df["type of gear"].replace({
    "auto": "automatic", 
    "manual": "manual", 
    "semi-automatic": "semi-automatic",
    "other": "other"
})

df["type of gear"] = df["type of gear"].fillna(df["type of gear"].mode()[0])
```

```python
df["type of gear"].value_counts()
```

```python
df['rating'] = df['rating'].abs()
```

```python
cleaned_df = df.copy()
cleaned_df.to_csv('CleanedData.csv', index=False)
```
