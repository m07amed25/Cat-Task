```python
import pandas as pd

df = pd.read_csv('CleanedData.csv')
```

```python
# Find the most expensive model for each brand
most_expensive_by_brand = df.loc[df.groupby('Brand')['price'].idxmax()][['Brand', 'VehicleModel', 'price']]
most_expensive_by_brand
```

```python
# Find the car type with the most cylinders
car_type_most_cylinders = df.loc[df['CylinderCount'].idxmax()][['Type', 'CylinderCount']]
car_type_most_cylinders
```

```python
# Find the average price per car type
average_price_by_type = df.groupby('Type')['price'].mean().reset_index()
average_price_by_type
```
